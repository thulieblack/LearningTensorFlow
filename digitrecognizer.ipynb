```python
import tensorflow as tf
import pandas as pd
import numpy as np
import keras
```

```python
test = pd.read_csv('../input/digit-recognizer/test.csv')
test.head()
```

```python
train = pd.read_csv('../input/digit-recognizer/train.csv')
train.head()
```

```python
train.shape
```

```python
y_train = train['label']
x_train = train.drop(labels=['label'],axis=1)
```

```python
#divide the data by 255.0 to normalize the dataset
x_train = x_train/255.0
test = test/255.0
```

```python
# split the dataset
from sklearn.model_selection import train_test_split
 
x_train,x_val,y_train,y_val =  train_test_split(x_train,y_train,test_size=0.1,random_state=2)
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(128,activation='relu'))
model.add(keras.layers.Dense(128,activation='relu'))
model.add(keras.layers.Dense(10,activation='softmax'))
```

```python
 model.compile(
    loss="sparse_categorical_crossentropy",
    optimizer = 'adam',
    metrics=['accuracy']
)
```

```python
hist = model.fit( x_train,y_train,validation_data=(x_val,y_val),epochs=10)
```

```python
pred = model.predict(test)
pred
```

```python
pred = np.argmax(pred,axis=1)
pred
```

```python
pred.shape
```

```python
submit = pd.read_csv('../input/digit-recognizer/sample_submission.csv')

#change the sample submission value to the predicted value
submit.Label = pred
```
